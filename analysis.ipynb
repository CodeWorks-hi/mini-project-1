```python
# 컬럼별 데이터 타입 확인 -> object는 문자열, int64는 정수
# 거래 금액은 문자열에서 정수로 바꾸어줄 필요
df.dtypes
```

```python
# 거래 금액 데이터 타입 변경
df["거래 금액 (Transaction Amount)"] = df["거래 금액 (Transaction Amount)"].str.replace(",", "").astype(int)
```

```python
df
```

```python
# 포인트 1. 휴대폰번호가 동일한 사람 존재 -> 중복 제거 필요?
df[df["휴대폰번호 (Phone Number)"].isin(df["휴대폰번호 (Phone Number)"].value_counts()[df["휴대폰번호 (Phone Number)"].value_counts() > 1].index)]
```

```python
# 포인트 2. 아이디가 동일한 사람 존재 -> 중복 제거 필요?
df[df["아이디 (User ID)"].isin(df["아이디 (User ID)"].value_counts()[df["아이디 (User ID)"].value_counts() > 1].index)]
```
