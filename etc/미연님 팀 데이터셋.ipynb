```python
df = df.rename(columns={
    "이름 (Name)": "이름",
    "생년월일 (Date of Birth)": "생년월일",
    "성별 (Gender)": "성별",
    "이메일 (Email)": "이메일",
    "휴대폰번호 (Phone Number)": "휴대폰번호",
    "주소 (Address)": "주소",
    "아이디 (User ID)": "아이디",
    "가입일 (Registration Date)": "가입일",
    "고객 세그먼트 (Customer Segment)": "등급",
    "차량구분(vehicle types)": "차 종",
    "구매한 제품 (Purchased Product)": "구매한 제품",
    "제품 구매 날짜 (Purchase Date)": "구매 날짜",
    "거래 금액 (Transaction Amount)": "금액",
    "거래 방식 (Transaction Method)": "거래 방식",
    "제품 구매 빈도 (Purchase Frequency)": "구매 횟수",
    "제품 구매 경로 (Purchase Path)": "구매 경로"
})
```

```python
# 현재 연도 기준으로 나이 계산
df["나이"] = df["생년월일"].apply(lambda x: datetime.now().year - int(x.split("-")[0]))
```

```python
# 연령대를 범주형 변수로 변환
bins = [0, 25, 35, 50, 65, 100]
labels = ["20대 이하", "26~35세", "36~50세", "51~65세", "66세 이상"]
df["연령대"] = pd.cut(df["나이"], bins=bins, labels=labels)
```

```python
print(df[["생년월일", "나이", "연령대"]].head())
```

```python
# 계절 구분 (봄: 3~5월, 여름: 6~8월, 가을: 9~11월, 겨울: 12~2월)
def get_season(date):
    month = int(date.split("-")[1])
    if month in [3, 4, 5]:
        return "봄"
    elif month in [6, 7, 8]:
        return "여름"
    elif month in [9, 10, 11]:
        return "가을"
    else:
        return "겨울"

df["구매 시즌"] = df["구매 날짜"].apply(get_season)

print(df[["구매 날짜", "구매 시즌"]].head())
```

```python
# 고객 등급을 점수로 변환 (VIP: 3, 일반: 2, 신규: 1, 이탈가능: 0)
grade_mapping = {"vip": 3, "일반": 2, "신규": 1, "이탈가능": 0}
df["고객 등급 점수"] = df["등급"].map(grade_mapping)
```

```python
df.head()
```

```python
def get_brand(model):
    if "Avante" in model or "Santa-Fe" in model or "Grandeur" in model:
        return "현대"
    elif "G70" in model or "G80" in model or "G90" in model:
        return "제네시스"
    elif "Tucson" in model or "NEXO" in model:
        return "기아"
    else:
        return "기타"

df["브랜드"] = df["구매한 제품"].apply(get_brand)


```

```python
print(df[["구매한 제품", "브랜드"]].head())
```

```python
segment_mapping = {
    "준중형 세단": "소형",
    "중형 세단": "중형",
    "준대형 세단": "중형",
    "대형 세단": "대형",
    "중형 suv": "SUV",
    "준중형 suv": "SUV",
    "대형 suv": "SUV",
    "픽업트럭": "트럭",
    "소형 해치백": "소형"
}
df["차급"] = df["차 종"].map(segment_mapping)


```

```python
print(df[["차 종", "차급"]].head())
```

```python
# 최근 1년 내 구매한 고객인지 여부 (True/False)
df["최근 구매 고객"] = df["구매 날짜"].apply(lambda x: int(x.split("-")[0]) >= 2023)


```

```python
print(df[["구매 날짜", "최근 구매 고객"]].head())
```

```python
# 누적 구매 금액
df["금액"] = df["금액"].str.replace(",", "").astype(int)
df["누적 구매 금액"] = df.groupby("주소")["금액"].cumsum()

# 평균 구매 금액
df["평균 구매 금액"] = df.groupby("주소")["금액"].transform("mean")

```

```python
print(df[["주소", "금액", "누적 구매 금액", "평균 구매 금액"]].head())
```

```python
grade_mapping = {"vip": 3, "일반": 2, "신규": 1, "이탈가능": 0}
df["고객 등급 점수"] = df["등급"].map(grade_mapping)
```

```python
df["구매 경로"] = df["구매 경로"].map({"온라인": 1, "오프라인": 0})
df["거래 방식"] = df["거래 방식"].map({"신용카드": 1, "계좌이체": 2, "현금": 3})
```

```python
df["누적 구매 금액"] = df.groupby("주소")["금액"].cumsum()
df["평균 구매 금액"] = df.groupby("주소")["금액"].transform("mean")
```

```python
#  "구매 요일" 추가 (구매 날짜 기반)
df["구매 요일"] = pd.to_datetime(df["구매 날짜"]).dt.day_name()
```

```python
#  "지역 카테고리" (주소에서 시/도를 기반으로 지역 그룹화)
df["지역 카테고리"] = df["주소"].apply(lambda x: x.split()[0] if pd.notna(x) else "기타")
```

```python
#  "고객 충성도 지수" (구매 횟수 및 누적 구매 금액을 기반으로 단순 지수화)
df["고객 충성도 지수"] = df["구매 횟수"] * df["누적 구매 금액"] / df["누적 구매 금액"].max()
```

```python
# 구매 날짜 변환
df["구매 날짜"] = pd.to_datetime(df["구매 날짜"], errors="coerce")
```

```python
# 할부 여부 (구매 금액 기준으로 설정, 4000만원 이상이면 할부 확률 높음)
df["할부 여부"] = df["금액"].apply(lambda x: 1 if x >= 40000000 else 0)
```

```python
# 마지막 구매 이후 기간 (월 기준)
df["마지막 구매 이후 기간"] = (datetime.now() - df["구매 날짜"]).dt.days / 30
```

```python
# 구매 주기 (각 고객별 구매 간격 계산)
df["구매 주기"] = df.groupby("주소")["구매 날짜"].diff().dt.days / 30
```

```python
# 구매 요일 카테고리 (주말 / 평일 구분)
df["구매 요일"] = df["구매 날짜"].dt.dayofweek  # 0(월) ~ 6(일)
df["구매 요일 카테고리"] = df["구매 요일"].apply(lambda x: "주말" if x >= 5 else "평일")
```

고객 평생 가치(Lifetime Value, LTV)는 한 고객이 기업과의 관계에서 얼마나 많은 가치를 창출하는지를 평가하는 지표입니다.
즉, "고객이 평생 동안 얼마를 소비할 것인가?" 를 예측하는 데 사용됩니다.

💡 공식 (LTV 계산)

고객 평생 가치 (LTV)
=
누적 구매 금액
×
고객 충성도 지수
고객 평생 가치 (LTV)=누적 구매 금액×고객 충성도 지수

```python
# 고객 평생 가치(LTV) = 누적 구매 금액 × 고객 충성도
df["고객 평생 가치"] = df["누적 구매 금액"] * df["고객 충성도 지수"]
```
